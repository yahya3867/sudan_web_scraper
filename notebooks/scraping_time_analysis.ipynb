```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('sudan-war.scraping-time.csv')

# Group by 'trial_number' and 'source' and calculate the mean duration
df_mean = df.groupby(['trial_number', 'source'], as_index=False)['duration'].mean()

# Create bar plot for the mean duration
fig = px.bar(
    df_mean,
    x="trial_number",
    y="duration",
    color="source",
    barmode='group',  # Group bars by Static and Dynamic
    labels={'trial_number': 'Trial Number', 'duration': 'Mean Duration'},
    title="Mean Duration for Static and Dynamic per Trial",
)

# Show the plot
fig.show()
```

```python
df = pd.read_csv('sudan-war.scraping-time.csv')

# Group by 'trial_number' and 'source' and calculate the mean duration
df_mean = df.groupby('source', as_index=False)['duration'].mean()

# Create bar plot for the mean duration
fig = px.bar(
    df_mean,
    x="source",
    y="duration",
    color="source",
    height=600,  # Set the height of the plot
    labels={'trial_number': 'Trial Number', 'duration': 'Mean Duration'},
    title="Mean Duration for Static and Dynamic per Trial",
)

# Show the plot
fig.show()
```

```python
df = pd.read_csv('sudan-war.scraping-time.csv')

# Group by 'trial_number' and 'source' and calculate the mean duration
df_mean = df.groupby(['source', 'trial_number'], as_index=False)['duration'].mean()

# Rename the source values
df_mean['source'] = df_mean['source'].replace('Static', 'Radio Tamazuj').replace('Dynamic', 'Sudan Tribune')

# Create a violin plot
fig = px.violin(
    df_mean,
    y='duration',          # Values to plot on the y-axis (duration)
    x='source',            # Categories to plot on the x-axis (Static and Dynamic)
    title="Distribution of Duration for Static and Dynamic Sources",
    labels={'duration': 'Duration (seconds)', 'source': 'Source'},
    category_orders={'source': ['Radio Tamazuj', 'Sudan Tribune']},
)

# Show the plot
fig.show()
```

```python
df = pd.read_csv('sudan-war.scraping-time.csv')

# Create a box plot
fig = px.box(
    df,
    y='duration',        # Y-axis is the 'duration'
    x='source',          # X-axis is 'source' (Static and Dynamic)
    title="Box Plot of Duration for Static and Dynamic Sources"
)

# Show the plot
fig.show()
```
